```python
import matplotlib.pyplot as plt

# Data for the three algorithms
algorithms = ['ADR', 'UCB', 'TS']
packet_delivery_ratio = [18.30, 54.80, 72.40]  # Replace with your actual values
average_current_consumption = [23.79, 16.89, 18.41]  # Replace with your actual values
colors = ['green', 'blue', 'orange']
alignment = 'center'

# Smaller bar width to minimize space between bars
bar_width = 1  # Adjust the width of the bars

# Function to add labels above each bar
def add_labels(ax, data, unit='%', fontsize=12):  # Specify the fontsize
    for i, val in enumerate(data):
        ax.text(i, val + 2, f'{val:.2f}{unit}', ha='center', va='bottom', fontsize=fontsize)
```

## Comparison of PDR of different algorithms

```python
# Plotting Packet Delivery Ratio with smaller bar width and no space between bars
fig, ax1 = plt.subplots(figsize=(5, 5))
ax1.bar(algorithms, packet_delivery_ratio, width=bar_width, color=colors, align=alignment)
ax1.set_xlabel('Algoritmus')
ax1.set_ylabel('Úspešnosť doručenia paketov [%]')
ax1.set_title('Porovnanie PDR pre rôzne algoritmy')
ax1.set_ylim(0, 100)  # Assuming Packet Delivery Ratio is between 0 and 100

# Adding labels above each bar
add_labels(ax1, packet_delivery_ratio, unit='%', fontsize=16)

# Save the chart as PNG
plt.savefig('packet_delivery_ratio_chart.png')

# Show the plots
plt.show()
```
